```python
def busca_regex(df, word):
    rgx="("+word+".*)"
    df['string_regex']=df['properati_url'].astype(str)+df['description'].astype(str)+df['title'].astype(str)
    df['string_regex']=df['string_regex'].str.lower()
    return df[df['string_regex'].str.count(rgx)==1]

def cuenta_regex(df, word):
    rgx="("+word+".*)"
    df['string_regex']=df['properati_url'].astype(str)+df['description'].astype(str)+df['title'].astype(str)
    df['string_regex']=df['string_regex'].str.lower()
    return df['string_regex'].str.count(rgx).sum()
```
